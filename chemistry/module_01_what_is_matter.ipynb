## 1.1 What is Matter?

**Everything you can touch, see, or feel is made of matter.**

Matter is anything that:
1. Has **mass** (it has "stuff" in it)
2. Takes up **space** (it occupies volume)

| Matter ✓ | NOT Matter ✗ |
|----------|-------------|
| A rock   | Light       |
| Water    | Sound       |
| Air (yes!)| Heat       |
| Your body | An idea    |

**Key insight:** Even things you can't see (like air) are matter — they have mass and take up space. You can feel wind push against you. That's air (matter) in motion.

### Distilled into Math

We said matter has **mass** and takes up **space** (volume). The relationship between these two properties gives us our first equation:

$$\rho = \frac{m}{V}$$

| Symbol | Name | Units | What it means |
|--------|------|-------|---------------|
| $\rho$ (rho) | **Density** | kg/m³ or g/cm³ | How tightly packed the matter is |
| $m$ | **Mass** | kg or g | How much "stuff" is there |
| $V$ | **Volume** | m³ or cm³ | How much space it takes up |

**Why it looks like this:** If you cram more mass into the same space, density goes up (m ↑, V same → ρ ↑). If you spread the same mass over more space, density goes down (m same, V ↑ → ρ ↓). It's a ratio — "how much stuff per unit of space."

**Intuition:** A brick and a sponge can be the same size (same V), but the brick has way more mass packed in → higher density.

> **Memorize:** $\rho = m / V$ — "density equals mass over volume"

## 1.2 What is Matter Made Of? — Atoms

If you zoom in on any piece of matter — say, a gold ring — you'd eventually reach the smallest unit: **atoms**.

An atom is so small that about **10 million** of them lined up side by side would span **1 millimeter**.

Let's visualize the zoom-in:

````python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# --- Panel 1: Visible gold ring ---
ax = axes[0]
ax.set_xlim(0, 10); ax.set_ylim(0, 10)
ring_outer = plt.Circle((5, 5), 3.5, color='#FFD700', ec='#B8860B', linewidth=3)
ring_inner = plt.Circle((5, 5), 2.2, color='white', ec='#B8860B', linewidth=2)
ax.add_patch(ring_outer)
ax.add_patch(ring_inner)
ax.set_title('1. What you see\n(a gold ring)', fontsize=13, fontweight='bold')
ax.set_aspect('equal')
ax.axis('off')

# --- Panel 2: Microscopic grains ---
ax = axes[1]
ax.set_xlim(0, 10); ax.set_ylim(0, 10)
np.random.seed(42)
for i in range(8):
    for j in range(8):
        x = 1.2 + i * 1.1 + np.random.uniform(-0.15, 0.15)
        y = 1.2 + j * 1.1 + np.random.uniform(-0.15, 0.15)
        color = plt.cm.YlOrBr(np.random.uniform(0.3, 0.6))
        grain = patches.RegularPolygon((x, y), 6, radius=0.5, color=color, ec='#8B7355', lw=0.5)
        ax.add_patch(grain)
ax.set_title('2. Zoom in (microscope)\n(crystal grains)', fontsize=13, fontweight='bold')
ax.set_aspect('equal')
ax.axis('off')

# --- Panel 3: Individual atoms ---
ax = axes[2]
ax.set_xlim(0, 10); ax.set_ylim(0, 10)
for i in range(7):
    for j in range(7):
        x = 1.5 + i * 1.15
        y = 1.5 + j * 1.15
        atom = plt.Circle((x, y), 0.45, color='#FFD700', ec='#B8860B', linewidth=1.5, alpha=0.9)
        ax.add_patch(atom)
        ax.text(x, y, 'Au', ha='center', va='center', fontsize=6, fontweight='bold', color='#5D4E37')
ax.set_title('3. Atomic level\n(individual gold atoms)', fontsize=13, fontweight='bold')
ax.set_aspect('equal')
ax.axis('off')

# Arrows between panels
fig.text(0.355, 0.5, '→ zoom →', ha='center', va='center', fontsize=16, color='gray')
fig.text(0.655, 0.5, '→ zoom →', ha='center', va='center', fontsize=16, color='gray')

plt.suptitle('Zooming into Matter', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
````

**Takeaway:** Everything — your desk, the ocean, the air — is made of atoms. They're just arranged differently.

## 1.3 Elements — The Building Blocks

An **element** is a substance made of only **one type** of atom.

- Gold → only gold atoms (Au)
- Oxygen → only oxygen atoms (O)
- Iron → only iron atoms (Fe)

There are about **118 known elements**. Everything in the universe is built from combinations of these 118 building blocks.

Think of elements like **LEGO brick types**:

````python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

colors_map = {'Au': '#FFD700', 'O': '#FF4444', 'Fe': '#A0A0A0', 'H': '#44AAFF'}

def draw_atoms(ax, atoms, title, subtitle):
    ax.set_xlim(0, 10); ax.set_ylim(0, 6)
    for i, (symbol, x, y) in enumerate(atoms):
        c = plt.Circle((x, y), 0.7, color=colors_map[symbol], ec='black', lw=1.5, alpha=0.85)
        ax.add_patch(c)
        ax.text(x, y, symbol, ha='center', va='center', fontsize=11, fontweight='bold')
    ax.set_title(f'{title}\n{subtitle}', fontsize=12, fontweight='bold')
    ax.set_aspect('equal')
    ax.axis('off')

# Pure element
gold_atoms = [('Au', 2, 3), ('Au', 3.7, 3), ('Au', 5.3, 3), ('Au', 7, 3), ('Au', 8.7, 3)]
draw_atoms(axes[0], gold_atoms, 'Pure Element: Gold', 'All the same atom type')

# Compound (water)
water = [('O', 3, 3), ('H', 1.5, 4), ('H', 4.5, 4),
         ('O', 7, 3), ('H', 5.5, 4), ('H', 8.5, 4)]
draw_atoms(axes[1], water, 'Compound: Water (H₂O)', 'Different atoms, chemically bonded')
# Draw bonds
for (x1,y1),(x2,y2) in [((3,3),(1.5,4)),((3,3),(4.5,4)),((7,3),(5.5,4)),((7,3),(8.5,4))]:
    axes[1].plot([x1,x2],[y1,y2], 'k-', lw=2, zorder=0)

# Mixture
mixture = [('Au', 1.5, 3), ('Fe', 3, 4), ('Au', 4.5, 2.5), ('O', 6, 3.5),
           ('Fe', 7.5, 3), ('Au', 9, 4)]
draw_atoms(axes[2], mixture, 'Mixture', 'Different atoms, NOT bonded')

plt.suptitle('Elements vs Compounds vs Mixtures', fontsize=15, fontweight='bold', y=1.05)
plt.tight_layout()
plt.show()
````

## 1.4 States of Matter

Matter exists in three common states. **The only difference between them is how much energy the particles have.**

| | Solid | Liquid | Gas |
|---|---|---|---|
| **Shape** | Fixed | Takes container shape | Fills everything |
| **Volume** | Fixed | Fixed | Expands to fill |
| **Particles** | Vibrate in place | Slide around | Fly freely |
| **Spacing** | Tight, ordered | Close, disordered | Far apart |
| **Energy** | Low | Medium | High |

Let's see this visually:

````python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
np.random.seed(7)

def draw_container(ax, title, color):
    ax.set_xlim(-1, 11); ax.set_ylim(-1, 11)
    # Container
    rect = patches.FancyBboxPatch((0, 0), 10, 10, boxstyle='round,pad=0.1',
                                   ec='#555', fc='#0a0a1a', lw=2)
    ax.add_patch(rect)
    ax.set_title(title, fontsize=14, fontweight='bold', color=color, pad=12)
    ax.set_aspect('equal')
    ax.axis('off')
    return ax

# --- SOLID ---
ax = draw_container(axes[0], 'SOLID', '#4fc3f7')
for i in range(6):
    for j in range(6):
        x = 1.5 + i * 1.4 + np.random.uniform(-0.1, 0.1)
        y = 1.5 + j * 1.4 + np.random.uniform(-0.1, 0.1)
        c = plt.Circle((x, y), 0.45, color='#4fc3f7', ec='white', lw=0.5, alpha=0.8)
        ax.add_patch(c)
# Draw grid bonds
for i in range(6):
    for j in range(6):
        x, y = 1.5 + i*1.4, 1.5 + j*1.4
        if i < 5: ax.plot([x, x+1.4], [y, y], color='#4fc3f7', alpha=0.2, lw=1)
        if j < 5: ax.plot([x, x], [y, y+1.4], color='#4fc3f7', alpha=0.2, lw=1)
ax.text(5, -0.7, 'Locked in place, vibrating', ha='center', fontsize=10, color='#888')

# --- LIQUID ---
ax = draw_container(axes[1], 'LIQUID', '#81c784')
for _ in range(36):
    x = np.random.uniform(1, 9)
    y = np.random.uniform(0.5, 6)  # mostly in bottom
    c = plt.Circle((x, y), 0.45, color='#81c784', ec='white', lw=0.5, alpha=0.8)
    ax.add_patch(c)
# "water level" line
ax.plot([0, 10], [6.5, 6.5], '--', color='#81c784', alpha=0.4, lw=1.5)
ax.text(5, -0.7, 'Sliding around, staying close', ha='center', fontsize=10, color='#888')

# --- GAS ---
ax = draw_container(axes[2], 'GAS', '#ef5350')
for _ in range(20):
    x = np.random.uniform(0.8, 9.2)
    y = np.random.uniform(0.8, 9.2)
    c = plt.Circle((x, y), 0.35, color='#ef5350', ec='white', lw=0.5, alpha=0.8)
    ax.add_patch(c)
    # Velocity arrows
    dx, dy = np.random.uniform(-1, 1), np.random.uniform(-1, 1)
    ax.annotate('', xy=(x+dx*0.6, y+dy*0.6), xytext=(x, y),
                arrowprops=dict(arrowstyle='->', color='#ef5350', lw=1.5, alpha=0.5))
ax.text(5, -0.7, 'Flying freely in all directions', ha='center', fontsize=10, color='#888')

plt.suptitle('Same Particles, Different Energy Levels', fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
````

### Phase Transitions

Adding or removing energy (heat) causes state changes:

```
  SOLID  ──heat──▶  LIQUID  ──heat──▶  GAS
         ◀──cool──          ◀──cool──
         
  melting/freezing      boiling/condensation
```

Let's plot what happens to temperature as you heat ice into steam:

### Distilled into Math — Heat & Temperature

The heating curve above shows that adding heat raises temperature. Here's the equation:

$$q = m \cdot c \cdot \Delta T$$

| Symbol | Name | Units | What it means |
|--------|------|-------|---------------|
| $q$ | **Heat** (energy transferred) | Joules (J) | How much energy you add or remove |
| $m$ | **Mass** | kg or g | How much stuff you're heating |
| $c$ | **Specific heat capacity** | J/(g·°C) | How "stubborn" a substance is to temperature change |
| $\Delta T$ | **Temperature change** | °C or K | $T_{final} - T_{initial}$ |

**Why it looks like this:**
- More mass ($m$ ↑) → need more heat to warm it all up
- Higher specific heat ($c$ ↑) → substance resists temperature change (water has high $c$, which is why oceans stabilize climate)
- Bigger temperature jump ($\Delta T$ ↑) → need more energy

**But what about phase changes?** During melting or boiling, temperature is *flat* — $\Delta T = 0$. So this equation gives $q = 0$, which is wrong! That's because during phase changes, energy goes into breaking bonds, not raising temperature. We use a different equation:

$$q = m \cdot L$$

| Symbol | Name | Units | What it means |
|--------|------|-------|---------------|
| $L$ | **Latent heat** | J/g | Energy needed to change state (per gram), without changing temperature |

- $L_f$ = latent heat of **fusion** (melting/freezing)
- $L_v$ = latent heat of **vaporization** (boiling/condensing)

**Why "latent"?** It means "hidden" — the heat is absorbed but you can't see it as a temperature change. It's hidden in the bond-breaking work.

> **Memorize:**
> - $q = mc\Delta T$ — heating within a phase (temperature changes)
> - $q = mL$ — changing between phases (temperature stays flat)

````python
# Heating curve for water: what happens when you continuously add heat to ice

fig, ax = plt.subplots(figsize=(12, 5))

# Heating curve segments (heat added vs temperature)
# Each phase: heat goes in, temp rises. At phase change: heat goes in, temp STAYS FLAT.
segments = [
    # (heat_start, heat_end, temp_start, temp_end, label, color)
    (0, 15, -20, 0, 'Ice warming', '#4fc3f7'),
    (15, 35, 0, 0, 'MELTING (ice → water)', '#00bcd4'),
    (35, 55, 0, 100, 'Water warming', '#81c784'),
    (55, 85, 100, 100, 'BOILING (water → steam)', '#ff9800'),
    (85, 100, 100, 130, 'Steam warming', '#ef5350'),
]

for h0, h1, t0, t1, label, color in segments:
    ax.plot([h0, h1], [t0, t1], color=color, linewidth=4, solid_capstyle='round')
    mid_h = (h0 + h1) / 2
    mid_t = (t0 + t1) / 2
    offset = 12 if t0 == t1 else 0
    ax.annotate(label, (mid_h, mid_t + offset), fontsize=9, ha='center',
                fontweight='bold', color=color,
                bbox=dict(boxstyle='round,pad=0.3', fc='#1a1a2e', ec=color, alpha=0.9))

# Key insight annotations
ax.annotate('Temperature STOPS rising\nduring phase changes!\n(energy breaks bonds instead)',
            xy=(25, 0), xytext=(25, -55),
            fontsize=10, ha='center', color='#00bcd4',
            arrowprops=dict(arrowstyle='->', color='#00bcd4', lw=2),
            bbox=dict(boxstyle='round,pad=0.5', fc='#0a0a1a', ec='#00bcd4'))

ax.set_xlabel('Heat Added →', fontsize=13, fontweight='bold')
ax.set_ylabel('Temperature (°C)', fontsize=13, fontweight='bold')
ax.set_title('Heating Curve of Water\n(What happens when you continuously add heat to ice)', 
             fontsize=14, fontweight='bold')
ax.set_facecolor('#0a0a1a')
fig.set_facecolor('#1a1a2e')
ax.tick_params(colors='#888')
ax.spines['bottom'].set_color('#555')
ax.spines['left'].set_color('#555')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_ylim(-70, 160)
ax.axhline(y=0, color='#555', linestyle=':', alpha=0.3)
ax.axhline(y=100, color='#555', linestyle=':', alpha=0.3)
ax.text(102, 0, '0°C (freezing)', fontsize=9, color='#888', va='center')
ax.text(102, 100, '100°C (boiling)', fontsize=9, color='#888', va='center')

plt.tight_layout()
plt.show()

print("""KEY INSIGHT:
During a phase change, temperature STOPS rising even though you keep adding heat.
Why? The energy is being used to break the bonds between particles,
not to make them move faster. Once all bonds are broken → temperature rises again.""")
````

## 1.5 Pure Substances vs. Mixtures

All matter falls into this classification tree:

````python
fig, ax = plt.subplots(figsize=(14, 8))
ax.set_xlim(0, 14); ax.set_ylim(0, 10)
ax.axis('off')
ax.set_facecolor('#0a0a1a')
fig.set_facecolor('#1a1a2e')

def box(ax, x, y, w, h, text, color, fontsize=11):
    rect = patches.FancyBboxPatch((x - w/2, y - h/2), w, h,
            boxstyle='round,pad=0.2', fc=color, ec='white', lw=1.5, alpha=0.9)
    ax.add_patch(rect)
    ax.text(x, y, text, ha='center', va='center', fontsize=fontsize,
            fontweight='bold', color='white')

def arrow(ax, x1, y1, x2, y2):
    ax.annotate('', xy=(x2, y2), xytext=(x1, y1),
                arrowprops=dict(arrowstyle='->', color='#888', lw=2))

# Tree structure
box(ax, 7, 9, 3, 1, 'MATTER', '#555', 14)

arrow(ax, 5.5, 8.5, 4, 7.3)
arrow(ax, 8.5, 8.5, 10, 7.3)

box(ax, 4, 6.8, 3.5, 1, 'Pure\nSubstance', '#1565C0')
box(ax, 10, 6.8, 3, 1, 'Mixture', '#2E7D32')

arrow(ax, 3, 6.3, 2, 5.0)
arrow(ax, 5, 6.3, 6, 5.0)
arrow(ax, 9, 6.3, 8, 5.0)
arrow(ax, 11, 6.3, 12, 5.0)

box(ax, 2, 4.5, 2.5, 1, 'Element', '#1976D2')
box(ax, 6, 4.5, 2.8, 1, 'Compound', '#0277BD')
box(ax, 8, 4.5, 3, 1, 'Homogeneous', '#388E3C')
box(ax, 12, 4.5, 3, 1, 'Heterogeneous', '#1B5E20')

# Examples below each
examples = [
    (2, 3.2, 'One type of atom\n\nGold (Au)\nOxygen (O₂)\nCarbon (C)', '#aaa'),
    (6, 3.0, '2+ elements bonded\nin fixed ratio\n\nWater (H₂O)\nSalt (NaCl)', '#aaa'),
    (8, 3.0, 'Uniform throughout\n(looks the same)\n\nSalt water\nAir', '#aaa'),
    (12, 3.0, 'Not uniform\n(see different parts)\n\nSalad\nSand + rocks', '#aaa'),
]
for x, y, text, color in examples:
    ax.text(x, y, text, ha='center', va='top', fontsize=8.5, color=color,
            style='italic', linespacing=1.4)

# Key distinction
ax.text(4, 1.0, '← Separated by CHEMICAL reactions', fontsize=10, color='#4fc3f7',
        ha='center', fontweight='bold')
ax.text(10, 1.0, 'Separated by PHYSICAL methods →', fontsize=10, color='#81c784',
        ha='center', fontweight='bold')
ax.text(7, 0.5, '(filtering, evaporating, etc.)', fontsize=9, color='#888', ha='center')

plt.title('Classification of Matter', fontsize=16, fontweight='bold', color='white', pad=15)
plt.tight_layout()
plt.show()
````
